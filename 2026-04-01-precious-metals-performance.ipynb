#### yfinance: comparing price of precious metals from 0ct-2025 to mar-2026

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Tickers
tickers = {
    "Gold": "GC=F",
    "Silver": "SI=F",
    "Copper": "HG=F",
    "Lithium (ETF)": "LIT"   # proxy
}

# Download 6 months data
raw = yf.download(list(tickers.values()), period="6mo")

# Use Close instead of Adj Close
data = raw["Close"]

# Rename columns
data.columns = tickers.keys()

# Normalize (optional, makes comparison easier)
normalized = data / data.iloc[0] * 100

# Plot
plt.figure(figsize=(12,6))
for col in normalized.columns:
    plt.plot(normalized.index, normalized[col], label=col)

plt.legend()
plt.title("Commodities (Last 6 Months, Normalized)")
plt.xlabel("Date")
plt.ylabel("Index (Base = 100)")
plt.grid()
plt.show()
```
