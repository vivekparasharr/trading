```python
%pip install yfinance mplfinance pandas pandas_ta matplotlib

import yfinance as yf
import pandas as pd
import pandas_ta as ta
import mplfinance as mpf
import matplotlib.pyplot as plt

# -----------------------------
# SETTINGS
# -----------------------------
ticker = "AAPL"        # change ticker
period = "1y"         # 6mo,1y,2y,5y
interval = "1d"

# -----------------------------
# DOWNLOAD DATA
# -----------------------------
df = yf.download(ticker, period=period, interval=interval, auto_adjust=True)

# flatten columns if needed
if isinstance(df.columns, pd.MultiIndex):
    df.columns = df.columns.get_level_values(0)

df.dropna(inplace=True)

# -----------------------------
# INDICATORS
# -----------------------------
# Trend
df["EMA20"] = ta.ema(df["Close"], length=20)
df["SMA50"] = ta.sma(df["Close"], length=50)

# Bollinger Bands
bb = ta.bbands(df["Close"], length=20, std=2)
df["BBL"] = bb["BBL_20_2.0"]
df["BBM"] = bb["BBM_20_2.0"]
df["BBU"] = bb["BBU_20_2.0"]

# RSI
df["RSI"] = ta.rsi(df["Close"], length=14)

# -----------------------------
# STYLE
# -----------------------------
mc = mpf.make_marketcolors(
    up='lime',
    down='red',
    edge='inherit',
    wick='inherit',
    volume='in'
)

style = mpf.make_mpf_style(
    base_mpf_style='nightclouds',
    marketcolors=mc,
    facecolor='#0d1117',
    figcolor='#0d1117',
    gridcolor='#2f3542',
    gridstyle='--',
    rc={'font.size': 10}
)

# -----------------------------
# ADDITIONAL PANELS
# -----------------------------
apds = [

    # Main panel overlays
    mpf.make_addplot(df["EMA20"], color='cyan', width=1.2),
    mpf.make_addplot(df["SMA50"], color='orange', width=1.2),

    mpf.make_addplot(df["BBU"], color='gray', width=0.8),
    mpf.make_addplot(df["BBM"], color='white', width=0.8),
    mpf.make_addplot(df["BBL"], color='gray', width=0.8),

    # RSI panel
    mpf.make_addplot(df["RSI"], panel=1, color='violet', ylabel='RSI', width=1.2),

    # RSI guide lines
    mpf.make_addplot([70]*len(df), panel=1, color='red', linestyle='--', width=0.8),
    mpf.make_addplot([30]*len(df), panel=1, color='lime', linestyle='--', width=0.8),
]

# -----------------------------
# PLOT
# -----------------------------
fig, axes = mpf.plot(
    df,
    type='candle',
    style=style,
    volume=True,
    addplot=apds,
    panel_ratios=(5, 2, 1),   # price, RSI, volume
    figsize=(16, 10),
    title=f"{ticker} Technical Dashboard",
    ylabel="Price",
    ylabel_lower="Volume",
    tight_layout=True,
    returnfig=True
)

plt.show()
```
